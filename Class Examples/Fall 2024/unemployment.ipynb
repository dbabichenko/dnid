### Deal with spaces in column names

```python
#df1.rename({'Country Code' : 'Country_Code'}, axis=1, inplace=True)
for col in df1.columns:
    df1.rename({str(col) : str(col).replace(' ', '_')}, axis=1, inplace=True)

for col in df2.columns:
    df2.rename({str(col) : str(col).replace(' ', '_')}, axis=1, inplace=True)
```

```python
df1.head()
```

### Merging DataFrames

```python
# Check if your key or your index is unique
df1['Country_Code'].unique()
```

```python
df1.shape
```

```python
len(df1['Country_Code'].unique())
```

```python
df1['Country_Code'].value_counts().sort_values(ascending=False)
```

```python
df = pd.merge(df1, df2, how='inner', left_on='Country_Code', right_on='Country_Code')
df.head()
```

### Sorting DataFrames

```python
df.sort_values('Country_Code') 
```

```python
df.sort_values('1997') 
```

```python
df.sort_values('1997', ascending=False) 
```

```python
df.sort_values(['1991', '2000', '1997'], ascending=False) 
```

```python
temp = df.sort_values(['1991', '2000', '1997'], ascending=False) 
```

```python
df.sort_values('Country_Code', ascending=False, inplace=True) 
```

```python
new_col_names = []
for col in df.columns:
    if str(col).strip().isnumeric():
        new_col_names.append("year_" + str(col))
    else:
        new_col_names.append(col)
        
df.columns = new_col_names
```

```python
df.head()
```
