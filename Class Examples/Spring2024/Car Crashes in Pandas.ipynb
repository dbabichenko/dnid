```python
# Rename column names to remove spaces
for col in df.columns:
    df.rename({col : col.replace(' ', '_')}, axis=1, inplace=True)

df.head()
```

```python
time_df = df.groupby('CRASH_TIME').agg({'NUMBER_OF_PEDESTRIANS_KILLED':'sum'})
time_df.reset_index(inplace=True)
time_df
```

```python
time_df.plot.scatter(x='CRASH_TIME', y='NUMBER_OF_PEDESTRIANS_KILLED')
```

```python
sns.scatterplot(data=time_df, x='CRASH_TIME', y='NUMBER_OF_PEDESTRIANS_KILLED')
```

```python
time_df.head()
```

```python
time_df['CRASH_TIME_NUM'] = 0
time_df.head()
```

```python
for idx, row in time_df.iterrows():
    crash_time = int(row['CRASH_TIME'].replace(':', ''))
    time_df.at[idx, 'CRASH_TIME_NUM'] = crash_time
```

```python
time_df.head()
```

```python
sns.scatterplot(data=time_df, x='CRASH_TIME_NUM', y='NUMBER_OF_PEDESTRIANS_KILLED')
```
